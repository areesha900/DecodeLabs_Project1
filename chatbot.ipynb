# DecodeLabs AI Internship - Project 1 : Rule-Based AI Chatbot

## Architecture: IPO Model

```
USER INPUT  ->  SANITIZATION  ->  DICTIONARY LOOKUP  ->  OUTPUT
   (raw)          (clean)            (O(1) match)        (reply)
```


### Knowledge Base (The Brain)

All chatbot rules live here as `key: value` pairs.
- **Key** = what the user types (always lowercase)
- **Value** = what the bot replies

````python
# KNOWLEDGE BASE

responses = {
    # Greetings
    'hello'         : 'Hey there! I am DBot. How can I help you today?',
    'hi'            : 'Hi! Welcome to DBot. What would you like to know?',
    'hey'           : 'Hey! Great to see you. Ask me anything!',

    # Small talk
    'how are you'        : 'I am running at 100% efficiency! No bad days for a bot.',
    'what is your name'  : 'I am DBot, your rule-based AI assistant built at DecodeLabs!',
    'who made you'       : 'I was built by an AI intern at DecodeLabs.',

    # AI knowledge
    'what is ai'                      : 'AI (Artificial Intelligence) is the simulation of human intelligence by machines.',
    'what is artificial intelligence' : 'Artificial Intelligence is the simulation of human intelligence by machines.',
    'what is ml'                      : 'ML (Machine Learning) is a subset of AI where systems learn from data instead of explicit rules.',
    'what is machine learning'        : 'Machine Learning is a subset of AI where systems learn from data instead of explicit rules.',
    'what is a chatbot'               : 'A chatbot is a program designed to simulate conversation with humans - like me!',
    'what is a rule based system'     : 'A rule-based system uses hard-coded logic or dictionaries to respond. No learning - just rules!',
    'difference between ai and ml'    : 'AI is the broad concept of smart machines. ML is one method to achieve AI through learning from data.',

    # DecodeLabs context
    'what is decodelabs' : 'DecodeLabs is the platform powering this AI internship. Visit www.decodelabs.tech!',
    'what is project 1'  : 'Project 1 is the Rule-Based AI Chatbot — a foundation milestone every intern must complete!',
    'what is ipo model'  : 'IPO stands for Input, Process, Output. It is the blueprint for transparent AI systems.',

    # Help
    'help' : 'I can answer: hello, how are you, what is ai, what is machine learning, what is decodelabs, what is project 1. Type exit to quit.',

    # Farewell
    'bye'       : 'Goodbye! Keep building. See you soon!',
    'goodbye'   : 'Goodbye! Keep building. See you soon!',
    'thank you' : 'You are welcome! Anything else I can help with?',
    'thanks'    : 'Happy to help! Type help to see what I know.',
}

EXIT_COMMANDS = {'exit', 'quit', 'bye', 'goodbye'}

print('Knowledge base loaded —', len(responses), 'intents ready.')
````

---
## Run the Chatbot

> When you run the cell, an input box will appear below. Type your message there.
>
> To stop the chatbot, type **exit**, **bye** or **quit**.

````python
# STARTUP MESSAGE
print('\n')
print('=' * 55)
print('DBot : Rule-Based AI Chatbot')
print('=' * 55)
print('\nBot: Hello! I am DBot.')
print('Bot: Type help to see what I can do, or exit to quit.\n')
print('-' * 55)
print('\n')


# HEARTBEAT LOOP
while True:

    # PHASE 1: INPUT & SANITIZATION
    raw_input = input('You: ')
    clean_input = raw_input.lower().strip()

    # Handle empty input
    if not clean_input:
        print('Bot: Please type something! Try help if you are stuck.')
        continue

    # EXIT STRATEGY
    if clean_input in EXIT_COMMANDS:
        print('Bot: Goodbye! Keep building!')
        print('=' * 55)
        break

    # PHASE 2: PROCESS
    reply = responses.get(clean_input, 'I do not understand that yet. Type help to see what I know!')

    # PHASE 3: OUTPUT
    print(f'Bot: {reply}')
    print()
````
